# Initialize libraries

```python
#initailize libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

# Import CSV File

```python

df = pd.read_csv('Rates_and_Trends_in_Heart_Disease_and_Stroke_Mortality_Among_US_Adults_35_by_CountyAge_GroupRace_Ethnicityand_Sex__2000-2019.csv')
df.head() #Here df as data frame as we can see below
```

```python
df.dtypes #data type of each column
```

```python
df.describe(include = 'object').T #specify each column in detail
```

# Feature Selection

```python
#choose columns where you want to work
selected_columns = ['Year','LocationDesc','Topic','Data_Value','Confidence_limit_Low',
                    'Confidence_limit_High','Stratification2','Stratification3','Stratification1']
df = df[selected_columns] #assign new variable
df
```

# Rename column

```python
#use dictionary techniqe to rename column
#here we use sorted_selected_df because we are constantly filtering the data frame and our last data frame name is sorted_selected_df. 
new_names = {'Topic': 'Disease_type','Data_Value': 'Death_population','Stratification2': 'Home_country',
             'Stratification3': 'sex','Stratification1': 'Age_group'}
df = df.rename(columns = new_names) #most important thing is if you place inplace = ture like(columns = new_names, inplace = True then this will modify existing table and you dont have to assign new one but if you dont do that then you are assign new table so keep in mind)
df
```

# Data Formating(place year in index format)


```python

# Convert the 'Year' column to a Pandas DateTimeIndex
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Set the 'Year' column as the index
df.set_index('Year', inplace=True)

# Display the updated DataFrame
df.head()
```

# Data filter according to  2018 and 2019 years

```python
# Filter rows for specific years 2018 and 2019 and assign it back to df
df = df[df.index.year.isin([2018, 2019])]
df

```

# check missing data

```python
df.isnull().sum()
```

# Handeling Missing Value

```python

# Fill NaN values in 'Death_population', 'Confidence_limit_Low', and 'Confidence_limit_High' columns with mode

# Calculate mode for 'Death_population', 'Confidence_limit_Low', and 'Confidence_limit_High' columns
mode_values = {
    'Death_population': df['Death_population'].mode()[0],
    'Confidence_limit_Low': df['Confidence_limit_Low'].mode()[0],
    'Confidence_limit_High': df['Confidence_limit_High'].mode()[0]
}

# Fill NaN values with mode using fillna() and the mode values dictionary
df.fillna(value=mode_values, inplace=True)

# Display the DataFrame
df

```

# Identify and remove outliers 

```python

# Statistical outlier detection using Z-score method

# Specify the columns for outlier detection
columns_for_outlier_detection = ['Death_population', 'Confidence_limit_Low', 'Confidence_limit_High']

# Loop through each column for outlier detection
for column in columns_for_outlier_detection:
    # Calculate Z-score for the current column
    z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
    
    # Define the threshold for outlier detection
    threshold = 3
    
    # Filter the DataFrame to exclude outliers for the current column
    df = df[z_scores <= threshold]

# Convert float columns to integer after outlier removal
df['Death_population'] = df['Death_population'].round().astype(int)
df['Confidence_limit_Low'] = df['Confidence_limit_Low'].round().astype(int)
df['Confidence_limit_High'] = df['Confidence_limit_High'].round().astype(int)

# Display the DataFrame after outlier removal and data type conversion
df

```

# Data Visualization


Pie Chart

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dfff = pd.DataFrame(df)

# Plotting
plt.figure(figsize=(10, 8))
average_by_year = dfff.groupby('Age_group')['Death_population'].mean()

plt.pie(average_by_year, labels=average_by_year.index, autopct='%1.1f%%', startangle=140)

# Plotting the line chart
average_by_year.plot(marker='o', linestyle='-')


plt.title('Death ratio according to Age Group')

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

death_total = df.groupby('sex')['Death_population'].mean().reset_index()
sns.set_style("whitegrid")
plt.figure(figsize=(10, 8))
sns.barplot(data=death_total, x='sex', y='Death_population', color='red', alpha=0.9)

# Set the title of the plot
plt.title('Average Death Population', fontsize=16)
# Set the label for the x-axis
plt.xlabel('Sex', fontsize=14)
# Set the label for the y-axis
plt.ylabel('Death Population', fontsize=14)
plt.show()
```
