```python
deaths_causes = df.iloc[:, 3:].sum().sort_values(ascending = False).head(10) #sum top 10 death cause and arrange in descending order
deaths_causes
```

# pie chart to compare death population of each country

```python
#pie chart to comapre deathcase
plt.figure(figsize=(5, 5)) #figure size,more you increase value smaller will be the size
explode = (0.1, 0.1) + (0,) * (len(deaths_causes) - 2) #split only two parts of piechar
plt.pie(deaths_causes, autopct='%1.1f%%', startangle=140, explode=explode, shadow = True) #plot with shadow in split
plt.title('Death Cases')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 

plt.legend(deaths_causes.index, loc="center left", bbox_to_anchor=(1, 0.5)) #legend position

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python
cardio_disease =df.groupby("Country/Territory")["Cardiovascular Diseases"].sum().sort_values(ascending = False).head(10)
cardio_disease
```

# bargraph to show top10 country with high death ratio

```python

plt.figure(figsize = (10,6))
sns.barplot(x= cardio_disease.index, y =cardio_disease)
plt.title('Death Case of Cardiovascular Disease by Country')
plt.xticks(rotation=45, ha='right') 
plt.xlabel('Countries')
plt.ylabel('Death Population')
plt.show()
```

# line chart to show time series death case of cardiovascular disease by country 

```python

for country in cardio_disease.index[:5] :    # pic top 4 countries from cardio_disease variable
    c_data = df[df['Country/Territory'] == country]
    sns.lineplot(x ='Year', y ='Cardiovascular Diseases', data = c_data, label = country, marker = 'o',markersize = 8 )
plt.title('Cardio by Country and Date')
plt.xlabel('Year')
plt.ylabel('Cardio death')
plt.legend(title = 'countries',bbox_to_anchor=(1.05,1), loc = 'upper left')
plt.show()
```

```python
neoplasm_disease = df.groupby("Country/Territory")["Neoplasms"].sum().sort_values(ascending = False).head(10)
neoplasm_disease
```

# line chart to show time series death case of Neoplasm disease by country 

```python
#Neoplasm by country and date
for country in neoplasm_disease.index[:10] :    # pic top 4 countries from cardio_disease variable
    c_data = df[df['Country/Territory'] == country]
    sns.lineplot(x ='Year', y ='Neoplasms', data = c_data, label = country, marker = 'o',markersize = 8 )
plt.title('Neoplasm by Country and Date')
plt.xlabel('Year')
plt.ylabel('Neoplasm death')
plt.legend(title = 'countries',bbox_to_anchor=(1.05,1), loc = 'upper left')
plt.show()
```
