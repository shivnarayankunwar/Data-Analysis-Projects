# Clean Wrong Data Entries and Flag IsRefund and IsPurchase in Original Data

```python
# Ensuring indices are aligned and exist in the original DataFrame
irrelevant_indices = Wrong_data_entries.index

# Drop irrelevant rows using their indices
data_partially_cleaned = data.drop(index=irrelevant_indices, errors='ignore')  # errors='ignore' prevents errors if indices are missing

# Flag refunds (InvoiceNo starts with 'C')
data_partially_cleaned['IsRefund'] = data_partially_cleaned['InvoiceNo'].str.startswith('C', na=False)

# Flag purchases (Quantity > 0)
data_partially_cleaned['IsPurchase'] = data_partially_cleaned['Quantity'] > 0
data_partially_cleaned
```

# Reformat InvoiceDate

```python
"""
since data has highly inconsistent formats,
so we use external libraries like dateutil.
parser for more flexible parsing.
"""
from dateutil.parser import parse
def flexible_date_parser(date):
    try:
        return parse(date)
    except:
        return None  # Return None for invalid entries

data_partially_cleaned['InvoiceDate'] = data_partially_cleaned['InvoiceDate'].apply(flexible_date_parser)
data_partially_cleaned
```

# Use NLP Technique To Categorized Descriptions

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Vectorize the descriptions
vectorizer = TfidfVectorizer(stop_words='english', max_features=2000)  # Adjust max_features for better results
X = vectorizer.fit_transform(data['Description'])

# Set the number of clusters
n_clusters = 15  # Adjust based on my dataset
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit and predict clusters
data['Category'] = kmeans.fit_predict(X)

for cluster in sorted(data['Category'].unique()):
    print(f"\nCluster {cluster}:")
    print(data[data['Category'] == cluster]['Description'].head(10))
```

# Consistent Revenue Growth Trend

```python
#Analyze Monthly Revenue Trends
# Add 'YearMonth' column for better practice
data['YearMonth'] = data['InvoiceDate'].dt.to_period('M')
category_monthly_revenue = data.groupby(['Category', 'YearMonth']).agg({
    'Total_revenue': 'sum'  # Monthly revenue for each category
}).reset_index()

# Calculate month-over-month revenue change for each category
category_monthly_revenue['RevenueChange'] = category_monthly_revenue.groupby('Category')['Total_revenue'].pct_change()

# Summarize positive and negative growth per category
growth_summary = category_monthly_revenue.groupby('Category').agg(
    TotalMonths=('YearMonth', 'count'),
    PositiveGrowth=('RevenueChange', lambda x: (x > 0).sum()),  # Count months with positive growth
    NegativeGrowth=('RevenueChange', lambda x: (x < 0).sum())   # Count months with negative growth
).reset_index()

# Calculate the percentage of months with positive growth
growth_summary['PositiveGrowthPercentage'] = (growth_summary['PositiveGrowth'] / growth_summary['TotalMonths']) * 100

# Identify categories with sustained growth (>60% positive growth)
sustained_growth_categories = growth_summary[growth_summary['PositiveGrowthPercentage'] > 60]

# Print categories with sustained growth
print("Categories with Sustained Revenue Growth:")
print(sustained_growth_categories[['Category', 'PositiveGrowthPercentage']])

# Plot Monthly Revenue Trends for Categories with Sustained Growth
if not sustained_growth_categories.empty:
    sustained_categories = sustained_growth_categories['Category']
    plt.figure(figsize=(16, 8))
    for category in sustained_categories:
        subset = category_monthly_revenue[category_monthly_revenue['Category'] == category]
        plt.plot(subset['YearMonth'].astype(str), subset['Total_revenue'], marker='o', label=category)

    plt.title('Consistent Revenue Growth Trend', fontsize=16, fontweight='bold')
    plt.xlabel('Year-Month', fontsize=12)
    plt.ylabel('Monthly Revenue', fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.legend(title='Category')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("No categories show sustained revenue growth based on the criteria.")
```

# Categories with Sustained Increased in Sell

```python

# Calculate monthly quantity sold for each category
category_monthly_quantity = data.groupby(['Category', 'YearMonth']).agg({
    'Quantity': 'sum'  # Monthly quantity sold for each category
}).reset_index()

# Calculate month-over-month percentage change for each category
category_monthly_quantity['QuantityChange'] = category_monthly_quantity.groupby('Category')['Quantity'].pct_change()

# Summarize the growth (positive and negative changes) for each category
growth_summary = category_monthly_quantity.groupby('Category').agg(
    TotalMonths=('YearMonth', 'count'),
    PositiveGrowth=('QuantityChange', lambda x: (x > 0).sum()),  # Count months with positive growth
    NegativeGrowth=('QuantityChange', lambda x: (x < 0).sum())   # Count months with negative growth
).reset_index()

# Calculate percentage of months with positive growth
growth_summary['PositiveGrowthPercentage'] = (growth_summary['PositiveGrowth'] / growth_summary['TotalMonths']) * 100

# Identify categories with sustained growth (>60% positive growth)
sustained_growth_categories = growth_summary[growth_summary['PositiveGrowthPercentage'] > 60]

# Print sustained growth categories
print("Categories with Sustained Increase in Quantity Sold:")
print(sustained_growth_categories[['Category', 'PositiveGrowthPercentage']])

# Visualize monthly quantity trends for sustained growth categories
if not sustained_growth_categories.empty:
    sustained_categories = sustained_growth_categories['Category']
    plt.figure(figsize=(16, 8))
    
    for category in sustained_categories:
        subset = category_monthly_quantity[category_monthly_quantity['Category'] == category]
        plt.plot(
            subset['YearMonth'].astype(str),
            subset['Quantity'],
            marker='o',
            label=category
        )

    # Add titles and labels
    plt.title('Categories with Sustained Increase in Quantity Sold', fontsize=16, fontweight='bold')
    plt.xlabel('Year-Month', fontsize=12)
    plt.ylabel('Quantity Sold', fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.legend(title='Category')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Show the plot
    plt.show()
else:
    print("No categories show sustained increase in quantity sold based on the criteria.")
```

# Solving Question Number 3 

```python
# Function to extract the season based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Add a 'Season' column to the dataset
data['Season'] = data['Invoice_months'].apply(get_season)

# Calculate total quantity sold for each category
category_quantity = data.groupby('Category')['Quantity'].sum().sort_values(ascending=False)

# Select the top 5 categories by quantity sold
top_5_categories = category_quantity.head(5).index

# Filter the data for the top 5 categories
top_categories_data = data[data['Category'].isin(top_5_categories)]

# Calculate seasonal quantity for the top 5 categories
seasonal_quantity_top_categories = (
    top_categories_data.groupby(['Season', 'Category'])['Quantity']
    .sum()
    .reset_index()
)

# Visualize the seasonal trends using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=seasonal_quantity_top_categories, 
    x='Season', 
    y='Quantity', 
    hue='Category', 
    palette='Set2'
)
plt.title('Seasonal Quantity Sold Trends for Top 5 Categories', fontsize=14)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Total Quantity Sold', fontsize=12)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

# Identifying Customer Recurrent pattern using  RFM Analysis

```python
#  Aggregate RFM Metrics
# Recency: Days since the last purchase
current_date = data_for_customer_behavior['InvoiceDate'].max()
rfm = data_for_customer_behavior.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (current_date - x.max()).days,  # Recency
    'InvoiceNo': 'count',  # Frequency
    'Total_revenue': 'sum'  # Monetary
}).reset_index()

# Rename columns
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'Total_revenue': 'Monetary'
}, inplace=True)

# Categorize RFM scores
# Define thresholds for high, medium, and low levels for each metric
rfm['Recency_Score'] = pd.qcut(rfm['Recency'], q=3, labels=[3, 2, 1])
rfm['Frequency_Score'] = pd.qcut(rfm['Frequency'], q=3, labels=[1, 2, 3])
rfm['Monetary_Score'] = pd.qcut(rfm['Monetary'], q=3, labels=[1, 2, 3])

# Combine RFM scores
rfm['RFM_Segment'] = rfm['Recency_Score'].astype(str) + rfm['Frequency_Score'].astype(str) + rfm['Monetary_Score'].astype(str)

# Visualize RFM Distribution
plt.figure(figsize=(15, 7))

# Recency Boxplot
plt.subplot(1, 3, 1)
sns.boxplot(y=rfm['Recency'], color='skyblue', width=0.5)
plt.title('Recency Distribution', fontsize=14, fontweight='bold')
plt.ylabel('Recency (days)', fontsize=12)
plt.xlabel('')

# Frequency Boxplot
plt.subplot(1, 3, 2)
sns.boxplot(y=rfm['Frequency'], color='orange', width=0.5)
plt.title('Frequency Distribution', fontsize=14, fontweight='bold')
plt.ylabel('Number of Purchases', fontsize=12)
plt.xlabel('')

# Monetary Boxplot
plt.subplot(1, 3, 3)
sns.boxplot(y=rfm['Monetary'], color='green', width=0.5)
plt.title('Monetary Distribution', fontsize=14, fontweight='bold')
plt.ylabel('Total Revenue', fontsize=12)
plt.xlabel('')

plt.tight_layout()
plt.show()
```
